```python
#import dependencies
import pandas as pd
```

```python
#Load data
nyc_bike = "202001-citibike-tripdata.csv"

bike_load = pd.read_csv(nyc_bike)
```

```python
#Calculate Age
bike_year = bike_load['birth year']

age_list = []

for y in bike_year:
    biker_age = 2020 - y
    age_list.append(biker_age)
```

```python
bike_load['gender'].replace([0,1,2],['Unknown','Male','Female'],inplace=True)

bike_load['age'] = age_list

bike_load
```

```python
groupby_user = bike_load.groupby('usertype')

groupby_user['usertype'].count()
```
